# 2. Library 1: NumPy (30 minutes)

## Key Concepts:

### Arrays: Faster and more efficient than Python lists.

## Operations: Arithmetic, slicing, and reshaping.

Example 1: Create and Manipulate Arrays

```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Perform operations
print("Add 10 to each element:", arr + 10)
print("Square each element:", arr ** 2)

# Reshape an array
matrix = np.arange(1, 10).reshape(3, 3)
print("3x3 Matrix:\n", matrix)
```

```python
# Example 2: Compute Statistics
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
print("Sum:", np.sum(data))
```

# 3. Library 2: Pandas (30 minutes)

## Key Concepts:

- DataFrame: A table-like data structure.

- Reading and Writing Data: CSV, Excel, etc.

- Data Selection and Manipulation.

- Example 1: Create a DataFrame

```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}
df = pd.DataFrame(data)
print(df)
```


# 4. Library 3: Matplotlib (30 minutes)

## Key Concepts:

- Line plots, bar plots, scatter plots.

- Customizing visualizations.

- Example 1: Create a Line Plot


```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, marker='o', linestyle='--', color='b')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Example 2: Create a Bar Plot

# Data
categories = ["A", "B", "C"]
values = [50, 70, 90]

plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# Example 3: Create a Scatter Plot

# Data
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 5. Hands-On Mini Project (30 minutes)

## Task: Analyze a dataset using Pandas, NumPy, and Matplotlib.

Problem: Analyze sales data to compute insights and visualize them.

Code:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample sales data
data = {
    "Product": ["A", "B", "C", "D", "E"],
    "Sales": [100, 200, 150, 300, 250],
    "Profit": [20, 50, 30, 80, 70]
}
sales_df = pd.DataFrame(data)

# Analyze Data
print("Total Sales:", sales_df["Sales"].sum())
print("Average Profit:", sales_df["Profit"].mean())

```

```python
# Visualize Data
plt.bar(sales_df["Product"], sales_df["Sales"], color="skyblue")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
